#정렬, Aggregation 함수, GroupBy 적용
데이터프레임과 Series의 정렬을 위해서는 sort_value() 매서드를 이용한다.
ascending = TRUE로 설정하면 오름차순으로 정렬(기본),
ascending = FALSE로 설정하면 내림차순으로 정렬한다.

input=True로 설정하면, 호출한 DataFrame의 정렬 결과를 그대로 적용합니다.
기본은 input=False입니다.

```python
# data_sfsnt 데이터를 FLO컬럼으로 오름차순에 의해 정렬해서 반환한다.

data_sfsnt_sorted = data_sfsnt.sort_values(by=["FLO"])
data_sfsnt_sorted.head(3)
```

```python
data_sfsnt_sorted = data_sfsnt.sort_values(by=["FLO", "STD"])
data_sfsnt_sorted.head(3)
```
